```python
# Preparacion rapida de datos
data.drop(columns=["Odd_1", "Odd_2"], inplace=True)
data['Winner_bool'] = data.apply(lambda x: 0 if x['Winner']==x['Player_1'] else 1, axis=1)
```

```python
# Extraer todos los nombres de los jugadores
all_players_p1_names = data['Player_1'].unique()
all_players_p2_names = data['Player_2'].unique()

a = set(all_players_p1_names)
b = set(all_players_p2_names)

all_players_names = list(b - a)
```

```python
# Explorar las estadisticas de cada jugador con la clase
from methods.explore_individual_player import PlayerStatictis

for player in  all_players_names:
    player_stats = PlayerStatictis(player)
```
